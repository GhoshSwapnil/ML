```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
#Binary Classification ( purchased ( 1 ) , not purchased ( 0 ) ) 
def analyzer(max_depth):
    data= pd.read_csv('dataset/Social_Network_Ads.csv')
    X= data.iloc[:, 2: 4].values
    Y= data.iloc[:, -1].values # purchased or not purchase 

    clf= DecisionTreeClassifier(max_depth= max_depth) # setting the max depth of the tree 
    clf.fit(X, Y)
    
    a= np.arange(start= X[:,0].min()- 1, stop= X[:,0].max()+ 1, step= 0.1)# age 
    b= np.arange(start= X[:,1].min()- 1, stop= X[:,1].max()+ 1, step= 100)# estimated salary 
    
    XX, YY= np.meshgrid(a, b)
    
    input_arr= np.array([XX.ravel(), YY.ravel()]).T
    
    label= clf.predict(input_arr)
    
    plt.contourf(XX, YY, label.reshape(XX.shape), alpha= 0.5) # continious data points predicted (classified) from original data points   
    plt.scatter(X[:, 0], X[:, 1], c=Y) # original data points     
```

```python
analyzer(1) # underfitting # depth 1 from root # very very less decision nodes # not able to properly classify
```

```python
analyzer(2) # near to optimal fit  # depth 2 from root   
```

```python
analyzer(4) # optimal fit # outliner not affecting so much # perfect classification # depth 4 from root   
```

```python
analyzer(5) # near to overfitting # depth 5 from root   
```

```python
analyzer(11) # overfitting # depth 11 from root # many decision nodes and large tree # smaller fragment # dosent works well in test data 
```

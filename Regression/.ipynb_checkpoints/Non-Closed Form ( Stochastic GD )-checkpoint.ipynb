```python
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, Y= load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size= 0.2,random_state= 2)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr= LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, lr.predict(X_test))
```

```python
class SGDRegressor:
    
    def __init__(self,learning_rate,epochs):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X_train,y_train):
        n= X_train.shape[0] # no of rows 
        m= X_train.shape[1] # number of columns 
        
        # b0= 0 # b1, b2,....bm= 1
        self.intercept_= 0
        self.coef_= np.ones(m).T# coefficient matrix 
        
        print("Intercept ( B0 ) -> ",self.intercept_,"  Coeffi ( b1, b2, b3, ....bm ) -> ", self.coef_.T)
         
        for i in range(self.epochs):
            for j in range(n):
                idx= np.random.randint(0, n) # random row index
                
                y_hat=  self.intercept_+ np.dot(X_train[idx], self.coef_)# current prediction(Y^) based on current coefficients
                
                intercept_der= -2*(y_train[idx] - y_hat) # intercept derevative # d(L)/d(b0) 
                self.intercept_= self.intercept_- (self.lr * intercept_der) # updtaing intercept ( b0 ) # applying GD 
                
                
                coef_der= -2 * np.dot( (y_train[idx]- y_hat), X_train[idx] ) # calculating all coefficient deravitive # d(L)/d(b1), d(L)/d(b2), d(L)/d(b3), ....d(L)/d(bm)   
                self.coef_= self.coef_ - (self.lr * coef_der) # updating all coefficient # b1, b2, b3, ....bm # applying GD
                 
                print("Intercept ( B0 ) -> ",self.intercept_,", Coeffi ( b1, b2, b3, ....bm ) -> ", self.coef_)
                
    def predict(self, X_test):
        return np.dot(X_test,self.coef_)+ self.intercept_
```

```python
sgdr = SGDRegressor(epochs=42,learning_rate=0.01)
```

```python
sgdr.fit(X_train, Y_train)
```

```python
r2_score(Y_test,sgdr.predict(X_test))
```

```python
class GDRegressor:
    
    def __init__(self,learning_rate,epochs):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X_train,y_train):
        n= X_train.shape[0] # no of rows 
        m= X_train.shape[1] # number of columns 
        
        # b0= 0 # b1, b2,....bm= 1
        self.intercept_= 0
        self.coef_= np.ones(m).T# coefficient matrix 
        
        print("Intercept ( B0 ) -> ",self.intercept_,"  Coeffi ( b1, b2, b3, ....bm ) -> ", self.coef_.T)
         
        for i in range(self.epochs):
            y_hat=  self.intercept_+ np.dot(X_train, self.coef_)# current prediction(Y^) based on current coefficients
            
            # updateing all the coef and the intercept using GD 
            
            intercept_der= -2 * np.mean(y_train - y_hat) # intercept derevative # d(L)/d(b0)
            self.intercept_= self.intercept_- (self.lr * intercept_der) # updtaing intercept ( b0 ) # applying GD 
            
            
            coef_der= (-2/ n)* np.dot( (y_train- y_hat).T, X_train ) # calculating all coefficient deravitive # d(L)/d(b1), d(L)/d(b2), d(L)/d(b3), ....d(L)/d(bm)   
            self.coef_= self.coef_ - (self.lr * coef_der) # updating all coefficient # b1, b2, b3, ....bm # applying GD
            
            print("Intercept ( B0 ) -> ",self.intercept_,", Coeffi ( b1, b2, b3, ....bm ) -> ", self.coef_) # printing coeff # how the coffecient are minimizing the error function and generating their value using Batch GD
    
    def predict(self, X_test):
        return np.dot(X_test,self.coef_)+ self.intercept_
```

```python
gdr = GDRegressor(epochs=1000,learning_rate=0.3)
```

```python
gdr.fit(X_train, Y_train)
```

```python
r2_score(Y_test,gdr.predict(X_test))
```

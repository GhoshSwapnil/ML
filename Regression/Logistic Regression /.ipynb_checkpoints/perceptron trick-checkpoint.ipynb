```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y= make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100) # linearly separable data
```

## Perceptron Trick 
- considering non-classified points in change the direction of classifcation line 
  - if missclassified point then pushing the classifcation line towards the point 
  - if classified point then doing nothing on the classifcation line 

```python
def perceptron(X,y, lr, epochs):
    
    X= np.insert(X,0,1,axis=1)# inserting one column of 1's as bias in the 0th position 
    n= X.shape[0] # no of rows 
    m= X.shape[1] # no of columns ( after inserting bias )
    
    coeff= np.ones(m) # w0,w1,w2= 1 # coefficient of starting line # 1* x1+ 1* x2+ 1* x0= 0 , <-> 1* x1+ 1* x2+ 1*1= 0
    lr= lr # learning rate 
    
    for i in range(epochs):
        idx= np.random.randint(0, n) # random row index from table 
        
        y_hat= step( np.dot( X[idx], coeff ) ) # 0 or 1 # our model predicted value 
        
        sign= y[idx] - y_hat # -1 means push line up, +1 push line down, 0 means line remains in same place  # is our algo predicted wrong then trying to correct the line it else keeping the line in same place  
        
        coeff= coeff+ lr * sign * X[idx]
        
    return coeff[0], coeff[1:]
```

```python
def step(z): # step function is a discreate function 
    if z >0 :
        return 1
    else:
        return 0
```

```python
intercept_,coef_ = perceptron(X,y, 0.1, 1000)
```

```python
print(coef_) # w1, w2
print(intercept_)# w0
```

```python
# ( w1* x1+ w2* x2+ w0* x0= 0 ) <-> ( w1* x1+ w2* x2+ w0= 0 ) # classification line equation 

m= -(coef_[0]/coef_[1]) # slope # m= -w2/w1
c= -(intercept_/coef_[1]) # intercept # c= -w0/w2
m, c
```

```python
x_input= np.linspace(-3,3, 100)# 100 points b/w 3 and -3 
y_input= m*x_input+ c
```

```python
x_input
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

## sklearn LogisticRegression

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3) # perceptron trick dosent worls so well on testing data as sklearn
plt.plot(x_input1,y_input1,color='yellow',linewidth=3) # sklearn line is more symmetric it performs well on testing data
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

## trying to go near sklearn classification using SIGMOID Function
- considering both classified and nonclassified point in changing the direction of classifcation line 
- to maintain the symmetry and equilibrium in line  
   - if classified point then repelling line in opposite direction 
   - if missclassified point the pushing the line towards the point 

```python
def Sperceptron(X,y, lr, epochs):
    
    X= np.insert(X,0,1,axis=1)# inserting one column of 1's as bias in the 0th position 
    n= X.shape[0] # no of rows 
    m= X.shape[1] # no of columns ( after inserting bias )
    
    coeff= np.ones(m) # w0,w1,w2= 1
    lr= lr # learning rate 
    
    for i in range(epochs):
        idx= np.random.randint(0, n) # random row index from table 
        
        y_hat= sigmoid( np.dot( X[idx], coeff ) ) # between 0 and 1 # our model predicted value 
        
        sign_eff= y[idx] - y_hat #sign and effect # -ve means push line up, +ve push line down # is our algo predicted wrong then trying to correct the line it else keeping the line in same place  
        
        coeff= coeff+ lr * sign_eff * X[idx]
        
    return coeff[0], coeff[1:]
```

```python
def sigmoid(z):
    return 1/(1+ np.exp(-z))
```

```python
intercept_,coef_ = Sperceptron(X,y, 0.1, 1000)
```

```python
print(coef_) # w1, w2
print(intercept_)# w0
```

```python
# ( w1* x1+ w2* x2+ w0* x0= 0 ) <-> ( w1* x1+ w2* x2+ w0= 0 ) # classification line equation 

m= -(coef_[0]/coef_[1]) # slope # m= -w2/w1
c= -(intercept_/coef_[1]) # intercept # c= -w0/w2
m, c
```

```python
x_input2= np.linspace(-3,3, 100)# 100 points b/w 3 and -3 
y_input2= m*x_input2+ c
```

```python
plt.figure(figsize=(10,6))

plt.plot(x_input,y_input,color='red',linewidth=3) # perceptron trick dosent worls so well on testing data as sklearn

plt.plot(x_input1,y_input1,color='yellow',linewidth=3) # sklearn line is symmetric it performs well on testing data

plt.plot(x_input2,y_input2,color='purple',linewidth=3) # percepton instead of step, sigmoid function # nearly symmetric like sklearn LogesticRegression

plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

## trying to go near sklearn classification using SIGMOID Function with batch Gradient Decent

```python
def gd(X,y, lr, epochs):
    
    X= np.insert(X,0,1,axis=1)# inserting one column of 1's as bias in the 0th position 
    n= X.shape[0] # no of rows 
    m= X.shape[1] # no of columns ( after inserting bias )
    
    weights= np.ones(m) # w0,w1,w2,...wm= 1
    lr= lr # learning rate 
    
    for i in range(epochs): #  all row at once batch GD # minmizing log loss error 
        y_hat= sigmoid( np.dot(X, weights) )
        weights= weights + lr * ( np.dot( (y-y_hat) , X) / n)
        
    return weights[0], weights[1:]
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
 intercept_, coef_= gd(X,y, 0.5, 5000)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input3= np.linspace(-3,3,100)
y_input3= m*x_input3 + b
```

```python
plt.figure(figsize=(10,6))

plt.plot(x_input,y_input,color='red',linewidth=3) # perceptron trick dosent worls so well on testing data as sklearn

plt.plot(x_input1,y_input1,color='yellow',linewidth=3) # sklearn line is symmetric it performs well on testing data

plt.plot(x_input2,y_input2,color='purple',linewidth=3) # percepton instead of step, sigmoid function # nearly symmetric like sklearn LogesticRegression

plt.plot(x_input3,y_input3,color='black',linewidth=3) #  sigmoid with GD works same as sklearn  

plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)

plt.ylim(-3,2)
```
